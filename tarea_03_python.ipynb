## 11. Función `domainGet`. Cree una función llamada domainGet que extraiga el dominio de un correo electrónico

Por ejemplo:
`domainGet("juan_perez@correo.com")` debería retornar 'correo.com'.

````python
# Copie su código acá
def domainGet(email):
  return email.split('@')[-1]

correo = "chachocoudet@correo.com"
print(domainGet(correo))
````

## 12. Función `buscar_palabra_promedio`

Cree una función llamada `palabra_promedio` que retorne `True` si la palabra "promedio" (en cualquier forma de mayúsculas/minúsculas) está contenida en el string de entrada.

Ejemplo:

`palabra_promedio("El Promedio de la clase es de 4.3")` debería retornar `True`.

````python
def palabra_promedio(frase):
  return "promedio" in frase.lower()

print(palabra_promedio("El Promedio del salon fue 3.8"))
````

## 13. Función contar_python

Cree una función llamada contar_python que cuente la cantidad de veces que aparece la palabra "python" en una cadena (ignorando mayúsculas y minúsculas).

Ejemplo:

`contar_python("Python es un lenguaje de programacion, Python es divertido")` debería retornar 2.

````python
# Copie su código acá
def contar_python(frase):
  return frase.lower().count('python')

texto = "Usar python es muy util para mejorar procesos"
result = contar_python(texto)
print(result)
````

## 14. 💣 Detección de anomalías

Un científico de datos necesita desarrollar una función para detectar anomalías en un conjunto de datos de valores positivos. La función `deteccion_anomalia` debe seguir estas reglas:

1. Recibe los siguientes argumentos:

    * `valor (float):` El valor a evaluar.

    * `valor_media (float):` La media del conjunto de datos.

    * `desviacion_estandar (float):` La desviación estándar.

    * `distribucion_normal (bool):` Indica si la distribución es normal.

2. Si la distribución es **normal** y el valor es mayor a **valor_media + 3 * desviacion_estandar**, la función retorna **"anormal"**.

3. Si la distribución es **normal** y el valor es menor a **valor_media - 3 * desviacion_estandar**, la función retorna **"normal"**.

4. En cualquier otro caso, retorna **"indeterminado"**.

````python
# Copie su código acá

def deteccion_anomalia(valor, valor_media, desviacion_estandar, distribucion_normal):
    limite_superior = valor_media + 3 * desviacion_estandar
    limite_inferior = valor_media - 3 * desviacion_estandar

    # Caso 1: Es normal y el valor es muy alto
    if distribucion_normal == True and valor > limite_superior:
        return "anormal"

    # Caso 2: Es normal y el valor es muy bajo
    elif distribucion_normal == True and valor < limite_inferior:
        return "normal"

    # Caso 3: Todo lo demás
    else:
        return "indeterminado"

resultado = deteccion_anomalia(50, 80, 20, True)

print(resultado)
````
