## 3. Cargar tablas a SQLite

```python
conn = sqlite3.connect(":memory:")

estudiantes.to_sql("estudiantes", conn, index=False, if_exists="replace")
cursos.to_sql("cursos", conn, index=False, if_exists="replace")
inscripciones.to_sql("inscripciones", conn, index=False, if_exists="replace")
pagos.to_sql("pagos", conn, index=False, if_exists="replace")

def run_sql(query):
    return pd.read_sql_query(query, conn)
```

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**

```python
# Punto 6 - SQL
# Escribe aquí tu consulta
query_6 = """
SELECT e.nombre, c.nombre_curso, i.estado
FROM estudiantes e
INNER JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
INNER JOIN cursos c ON i.curso_id = c.curso_id
"""
sql_6 = run_sql(query_6)
sql_6
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
# Escribe aquí tu consulta
query_8 = """
SELECT e.estudiante_id, e.nombre, COUNT(i.inscripcion_id) AS n_inscripciones
FROM estudiantes e
JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
GROUP BY e.estudiante_id, e.nombre
HAVING n_inscripciones >= 2
"""
sql_8 = run_sql(query_8)
sql_8
```

```python
# Punto 8 - pandas
# Escribe aquí tu solución
punto_8 = estudiantes.merge(inscripciones, on="estudiante_id")
punto_8 = punto_8.groupby(["estudiante_id", "nombre"]).size().reset_index(name="n_inscripciones")
punto_8[punto_8["n_inscripciones"] >= 2]
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
# Punto 9 - SQL
# Escribe aquí tu consulta
query_9 = """
SELECT
    nombre_curso,
    precio,
    CASE
        WHEN precio < 120 THEN 'Bajo'
        WHEN precio <= 180 THEN 'Medio'
        ELSE 'Alto'
    END AS segmento_precio
FROM cursos
"""
sql_9 = run_sql(query_9)
sql_9
```

```python
# Punto 9 - pandas
# Escribe aquí tu solución
import numpy as np

punto_9 = cursos[["nombre_curso", "precio"]].copy()
condiciones = [
    (punto_9["precio"] < 120),
    (punto_9["precio"] >= 120) & (punto_9["precio"] <= 180),
    (punto_9["precio"] > 180)
]
segmentos = ["Bajo", "Medio", "Alto"]
punto_9["segmento_precio"] = np.select(condiciones, segmentos, default="No definido")
punto_9
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
# Punto 10 - SQL
# Escribe aquí tu consulta

query_10 = """
SELECT
    c.categoria,
    SUM(p.monto) AS total_pagado
FROM cursos c
JOIN inscripciones i ON c.curso_id = i.curso_id
JOIN pagos p ON i.inscripcion_id = p.inscripcion_id
GROUP BY c.categoria
ORDER BY total_pagado DESC
"""
sql_10 = run_sql(query_10)
sql_10
```

```python
# Punto 10 - pandas
# Escribe aquí tu solución
punto_10 = (
    cursos.merge(inscripciones, on="curso_id")
    .merge(pagos, on="inscripcion_id")
    .groupby("categoria", as_index=False)["monto"]
    .sum()
    .rename(columns={"monto": "total_pagado"})
    .sort_values("total_pagado", ascending=False)
    .reset_index(drop=True)
)
punto_10
```
